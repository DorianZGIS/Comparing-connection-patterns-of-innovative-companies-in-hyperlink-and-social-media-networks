```python
comp_ids_needed = list(all15[all15.Twitter_handle.isin(nodes_graph)].comp_id)
```

```python
mean_distances_list = []
counter =0
distance = []
for c_id in comp_ids_needed:
    if int(c_id) in [int(x) for x in distances_lookup.keys()]:
        dist = (distances_lookup[int(c_id)])
        number = 0
        sum_distances = 0
        try:
            for d in dist:
                number += 1
                sum_distances += d
            mean_distance = sum_distances/number
            mean_distances_list.append(mean_distance)
            distance.append(dist)
        except:
            print(dist)
            counter+=1
            mean_distances_list.append(np.nan)
            distance.append(np.nan)
            # was nan so no distance was calculated, maybe because no friends
            
    else:
        #had no usertimeline
        mean_distances_list.append(np.nan)
        distance.append(np.nan)
        print(c_id)
        counter +=1
```
