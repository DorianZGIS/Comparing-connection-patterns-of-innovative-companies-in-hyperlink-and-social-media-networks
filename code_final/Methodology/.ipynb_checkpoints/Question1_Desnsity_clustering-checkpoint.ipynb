```python
try:
    import cPickle as pickle
except ImportError:  
    import pickle
with open('/Users/dorianbenutzer/Desktop/Digital_Twitter_Layer_local/data/graphs/pos_dic2_hyp.p', 'rb') as fp:
    pos_dic = pickle.load(fp)

# color by path length from node near center
p = clustering_hyp
G = weighted_directed_hyplinks_graph
plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(G, pos_dic,  alpha=0.4)
nx.draw_networkx_nodes(
    G,
    pos_dic,
    nodelist=list(p.keys()),
    node_size=60,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
    alpha =0.4
)


plt.axis("off")
plt.show()
```
