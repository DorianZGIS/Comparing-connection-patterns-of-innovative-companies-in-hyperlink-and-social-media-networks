```python
def create_pos_dict(network, data):
    df = pd.DataFrame(columns = ['Latitude', 'Longitude'])
    counter = 0
    for company_abi in network.ABI:
        counter += 1
        df = df.append(data[data['ABI']==company_abi][['Latitude', 'Longitude']])
        print('We are at company: ', counter)
    keys = network.company_names
    pos_list = [[df['Longitude'][ind],df['Latitude'][ind]]for ind in df.index]
    values = pos_list
    dic_pos = dict(zip(keys,values))
    return dic_pos
```

```python
def plot_map(p, G,path, normalize =True, factor = 1000):
    node_sizes = []
    # normalize with q75
    q75 = np.quantile(list(p.values()), 0.75)
    if normalize:
        maxx = max(p.values())
        for key in p.keys():
            val = p[key]
            if val > q75:
                p[key] = (val-q75)/(maxx-q75)

    alphas = []
    nodes = []
    colors = {}
    counter =0
    for n in G.nodes():
        val = p[n]
        colors[n] = val
        if val <= q75:
            alphas.append(0.1)
            node_sizes.append(5)
        else:
            alphas.append(0.9)
            node_sizes.append(val*factor)
            counter += 1
    print(counter)


    plt.figure(figsize=(15,20))
    m = Basemap(
    projection = 'cyl',
    llcrnrlon=-180,
    llcrnrlat=10,
    urcrnrlon=-50,
    urcrnrlat=70,
    lat_ts=0,
    resolution='l',
    suppress_ticks=True)


    nx.draw_networkx_edges(G, pos_dic,  alpha=0.1, edge_color ='#A9A9A9',arrows =False)
    
    nc = nx.draw_networkx_nodes(
        G,
        pos_dic,
        nodelist=G.nodes(),
        node_size = node_sizes,
        node_color=list(colors.values()),

        alpha = alphas
    )
    plt.axis("off")

    m.drawcountries(linewidth = 1)
    m.drawstates(linewidth = 0.2)
    m.drawcoastlines(linewidth=1)
    plt.tight_layout()
    plt.savefig(path, format = "png", dpi = 300)
    plt.tight_layout()
    plt.show()





```
