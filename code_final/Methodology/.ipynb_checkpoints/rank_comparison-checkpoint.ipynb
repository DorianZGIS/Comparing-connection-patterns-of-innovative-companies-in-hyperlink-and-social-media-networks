```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap as Basemap
```

```python
def plot_map(p, G,path, save, normalize =True, factor = 1000):
    node_sizes = []
    # normalize with q75
    p = dict(sorted(p.items(), key=lambda item: item[1]))
    q75 = np.quantile(list(p.values()), 0.75)
    if normalize:
        maxx = max(p.values())
        for key in p.keys():
            val = p[key]
            if val > q75:
                p[key] = (val-q75)/(maxx-q75)
    alphas = []
    nodes = []
    colors = {}
    counter = 0
    #for n in p.keys():
    #    val = p[n]
    #    colors[n] = val
    #    if val <= q75:
    #        alphas.append(0.1)
    #        node_sizes.append(5)
    #    else:
    #        alphas.append(0.9)
    #        node_sizes.append(val*factor)
    #        counter += 1
    #print('%d companies have values higher then q75'%(counter))
    
    plt.figure(figsize=(15,20))
    m = Basemap(
    projection = 'cyl',
    llcrnrlon=-180,
    llcrnrlat=10,
    urcrnrlon=-50,
    urcrnrlat=70,
    lat_ts=0,
    resolution='l',
    suppress_ticks=True)
    
    
    nx.draw_networkx_edges(G, pos_dic,  alpha=0.1, edge_color ='#A9A9A9',arrows =False)
    nc = nx.draw_networkx_nodes(
        G,
        pos_dic,
        nodelist=p.keys(),
        node_size = 10,
        node_color=list(p.values()),
        alpha = 1
    )
    plt.axis("off")
    m.drawcountries(linewidth = 1)
    m.drawstates(linewidth = 0.2)
    m.drawcoastlines(linewidth=1)
    plt.tight_layout()
    if save:
        plt.savefig(path, format = "png", dpi = 300)
    plt.tight_layout()
    plt.show()
```

```python
ranks = {}
G = weighted_directed_hyplinks_graph
rank_diff = rank1-rank2
counter = 0
for n in close['Unnamed: 0']:
    ranks[n] = rank_diff[counter]
    counter +=1
    
```

```python
p = ranks
maxx = max(p.values())
minn = min(p.values())
for key in p.keys():
    val = p[key]
    p[key] = (val-minn)/(maxx-minn)
```

```python
####################
# Hyperlinks closeness
####################
p = p
G = weighted_directed_hyplinks_graph
path = "/Users/dorianbenutzer/Desktop/Masterthesis/Scripts/results/geolocated_graph/in_closeness_hyp.png"
plot_map(p, G,path,save=False, normalize =False, factor = 100)
```
