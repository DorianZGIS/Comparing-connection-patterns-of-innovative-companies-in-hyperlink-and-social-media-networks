```python
def norm(v):
    maxx = max(v)
    minn = min(v)
    norm_v = []
    for val in v:
        norm_val=(val-minn)/(maxx-minn)
        norm_v.append(norm_val)
    return norm_v
```

https://tuangauss.github.io/projects/networkx_basemap/networkx_basemap.html

```python
def plot_map(p, G,path,save, normalize =True, factor = 1000):
    node_sizes = []
    # normalize with q75
    p = dict(sorted(p.items(), key=lambda item: item[1]))
    q75 = np.quantile(list(p.values()), 0.75)
    
    if normalize:
        maxx = max(p.values())
        for key in p.keys():
            val = p[key]
            if val > q75:
                p[key] = (val-q75)/(maxx-q75)
    else:
        minn = min(p.values())
        maxx = max(p.values())
        for key in p.keys():
            val = p[key]
            p[key] = (val-minn)/(maxx-minn)
        q75 = minn
        

    
    alphas = []
    nodes = []
    colors = {}
    counter =0
    
    for n in p.keys():
        val = p[n]
        colors[n] = val
        if val <= q75:
            alphas.append(0.3)
            node_sizes.append(5)
        else:
            alphas.append(0.9)
            node_sizes.append(val*factor)
            counter += 1
            
    print(counter)
    plt.figure(figsize=(15,20))
    m = Basemap(
    projection = 'cyl',
    llcrnrlon=-180,
    llcrnrlat=10,
    urcrnrlon=-50,
    urcrnrlat=70,
    lat_ts=0,
    resolution='l',
    suppress_ticks=True)
    
    nx.draw_networkx_edges(G, pos_dic, nodelist= p.keys(), alpha=0.1, edge_color ='#A9A9A9',arrows =False)
    nc = nx.draw_networkx_nodes(
        G,
        pos_dic,
        nodelist= p.keys(),
        node_size = node_sizes,
        node_color=list(colors.values()),
        alpha = alphas
    )
    plt.axis("off")
    m.drawcountries(linewidth = 1)
    m.drawstates(linewidth = 0.2)
    m.drawcoastlines(linewidth=1)
    plt.tight_layout()
    if save:
        plt.savefig(path, format = "png", dpi = 300)
    plt.tight_layout()
    plt.show()
```
