# Mean geographic proximity

```python
haversine_matrix = pd.read_csv('/Users/dorianbenutzer/Desktop/Digital_Twitter_Layer_local/data/graphs/haversine_matrix_with_names.csv')
# The distance is in km
```

### Create an edge dictionary

```python
###
# Hyperlink graph
###
edges = list(weighted_directed_hyplinks_graph.edges())
#nodes = list(weighted_directed_hyplinks_graph.nodes())
edges_dic_h = {}

for edge in edges:
    key = edge[0]
    val = edge[1]
    if key not in edges_dic_h.keys():
        edges_dic_h[key] = val
    else:
        val_old = edges_dic_h[key]
        val_new = val_old + ',' + val
        edges_dic_h[key] = val_new
```

```python
###
# Followers graph
###

edges = list(weighted_directed_followers_graph.edges())
#nodes = list(weighted_directed_followers_graph.nodes())
edges_dic_f = {}

for edge in edges:
    key = edge[0]
    val = edge[1]
    if key not in edges_dic_f.keys():
        edges_dic_f[key] = val
    else:
        val_old = edges_dic_f[key]
        val_new = val_old + ',' + val
        edges_dic_f[key] = val_new
```

```python
print('Length Follower graph edge dictionary: ', len(edges_dic_f.keys()))
print('Length Hyperlink graph edge dictionary: ',len(edges_dic_h.keys()))
```

### Create mean_geographical_proximity_dic

```python
haversine_matrix = haversine_matrix.drop('Unnamed: 0', axis=1)
haversine_matrix.index = haversine_matrix.columns
```

```python
###
# Hyperlink graph
###

mean_geographical_proximity_dic_hyp = {}

for comp in edges_dic_h.keys():
    comp_edge_list = edges_dic_h[comp].split(',')
    number = 0
    sum_distance = 0
    for edge in comp_edge_list:
        sum_distance += haversine_matrix.loc[comp,edge]
        number += 1
    mean_distance = sum_distance/number
    mean_geographical_proximity_dic_hyp[comp] = mean_distance
```

```python
###
# Followers graph 
###

mean_geographical_proximity_dic_foll = {}

for comp in edges_dic_f.keys():
    comp_edge_list = edges_dic_f[comp].split(',')
    number = 0
    sum_distance = 0
    for edge in comp_edge_list:
        sum_distance += haversine_matrix.loc[comp,edge]
        number += 1
    mean_distance = sum_distance/number
    mean_geographical_proximity_dic_foll[comp] = mean_distance
```

```python
print('Maximum distance bridged by follower connections: (in km)',max(mean_geographical_proximity_dic_foll.values()))
```

```python
print('Maximum distance bridged by hyperlinks: (in km) ',max( mean_geographical_proximity_dic_hyp.values()))
```

### Save in dataframe

```python
mgph = []
nodes = list(weighted_directed_followers_graph.nodes())
for node in nodes:
    if node in mean_geographical_proximity_dic_hyp.keys():
        val = mean_geographical_proximity_dic_hyp[node]
    else:
        val = 0 
    mgph.append(val)
mgpf = []
for node in nodes:
    if node in mean_geographical_proximity_dic_foll.keys():
        val = mean_geographical_proximity_dic_foll[node]
    else:
        val = 0
    mgpf.append(val)
```

```python
def norm(v):
    maxx = max(v)
    minn = min(v)
    norm_v = []
    for val in v:
        norm_val=(val-minn)/(maxx-minn)
        norm_v.append(norm_val)
    return norm_v

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

x = list(norm(np.log(geo.mean_geographical_proximity_hyp+1)))
y = list(norm(np.log(geo.mean_geographical_proximity_foll+1)))
data = pd.DataFrame({
    '(log) normalized mean geographical proximity hyperlink network': x,
    '(log) normalized mean geographical proximity follower network': y,
})
h = sns.jointplot(x='(log) normalized mean geographical proximity hyperlink network', 
                  y='(log) normalized mean geographical proximity follower network',
              data=data,color="#4CB391",kind="reg",ylim = (-0.1,1.1),xlim = (-0.1,1.1))
path = '/Users/dorianbenutzer/Desktop/Masterthesis/Scripts/results/Q3/geographic_proxi_log'
plt.savefig(path, format = "png", dpi = 300)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import re
from scipy.spatial.distance import cosine
```

```python
data["text"] = data["text"].apply(lambda x: re.sub("[^abcdefghijklmnopqrstuvwxyzäöüß&. ']", '', str(x).lower()).strip())
```

```python
new_data = data
print(len(new_data))
```

```python
tfidf_vectorizer = TfidfVectorizer(analyzer='word', max_features=None, min_df=0.015, max_df=0.65)
tfidf_vectors = tfidf_vectorizer.fit_transform(data["text"].values.tolist())
```

```python
tfidf_vectors
```

```python
## Remember the keys are the company ids

cog_distances_lookup = {}
start = time.time()
s2 = time.time()
data = new_data
comp_id_list = list(data.comp_id)
exceptions = []

for ind, row in data.iterrows():
    comp_id = row['comp_id']
    twitter_handle = row.Twitter_handle
    if row.edges!='[]' and '@' not in row.edges:
        friends_ids = row['edges'].replace('[','').replace(']','').split(', ')
        friends_ids_int = [int(x) for x in friends_ids]
        
        tfidf = tfidf_vectors[ind].todense()
        distances = []
        
        for friends_id in friends_ids_int:
            if friends_id in comp_id_list:
                friends_id_index = comp_id_list.index(friends_id)
                tfidf_partner = tfidf_vectors[friends_id_index].todense()
                try:
                    distance = cosine(tfidf, tfidf_partner)
                except Exception as exc:
                    exceptions.append((comp_id,exc))
                    
                distances.append(distance)
        if distances == []:
            distances = ("no_friends_in_sample")
        cog_distances_lookup[twitter_handle] = distances
    else:
        cog_distances_lookup[twitter_handle] = ("no_edges")
    if time.time()-s2 > 60:
        s2 = time.time()
        print(ind)
        
print(time.time()-start)
```

```python
# Remember the keys are the abi

data1 = data
node_list = list(data1.nodes)
exceptions = []
cog_distances_lookup_h = {}
start = time.time()
s2 = time.time()

for ind, row in data1.iterrows():
    edges = row.edges
    abi = row.ABI
    twitter_handle = row.company_names
    if pd.notnull(edges) == True: 
        edges_list = edges.split(',')
        tfidf = tfidf_vectors[ind].todense()
        if np.any(tfidf) != 0:
            distances = []
            for edge in edges_list:
                if edge in (node_list):
                    edge_index = node_list.index(edge)
                    tfidf_partner = tfidf_vectors[edge_index].todense()
                    if np.any(tfidf_partner) != 0:
                        try:
                            distance = cosine(tfidf, tfidf_partner)

                        except Exception as exc:
                            exceptions.append((ind,exc))
                            distances = ("exception_occured")
                    #else: #ignore that edge it has no text!
                    distances.append(distance)
            if distances == []:
                distances = ("no_friends_in_sample")
            cog_distances_lookup_h[twitter_handle] = distances
        else:
            cog_distances_lookup_h[twitter_handle] = ("zero_tfidf")
        
    else:
        cog_distances_lookup_h[twitter_handle] = ("no_edges")
    if time.time()-s2 > 10:
        s2 = time.time()
        print(ind)
        
print(time.time()-start)
```

```python
import networkx as nx
import time
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap as Basemap

def plot_map(p, G,path,save, normalize =True, factor = 1000):
    node_sizes = []
    # normalize with q75
    p = dict(sorted(p.items(), key=lambda item: item[1]))
    q75 = np.quantile(list(p.values()), 0.75)
    
    
    if normalize:
        maxx = max(p.values())
        for key in p.keys():
            val = p[key]
            if val > q75:
                p[key] = (val-q75)/(maxx-q75)
                
    else:
        minn = np.log(min(p.values())+1)
        maxx = np.log(max(p.values())+1)
        for key in p.keys():
            val = np.log(p[key]+1)
            p[key] = (val-minn)/(maxx-minn)
        q75 = minn

    
    alphas = []
    nodes = []
    colors = {}
    counter =0
    
    for n in p.keys():
        val = p[n]
        colors[n] = val
        if val <= q75:
            alphas.append(0.3)
            node_sizes.append(5)
        else:
            alphas.append(0.9)
            node_sizes.append(val*factor)
            counter += 1
            
    print(counter)
    plt.figure(figsize=(15,20))
    m = Basemap(
    projection = 'cyl',
    llcrnrlon=-180,
    llcrnrlat=10,
    urcrnrlon=-50,
    urcrnrlat=70,
    lat_ts=0,
    resolution='l',
    suppress_ticks=True)
    
    nx.draw_networkx_edges(G, pos_dic, nodelist= p.keys(), alpha=0.1, edge_color ='#A9A9A9',arrows =False)
    nc = nx.draw_networkx_nodes(
        G,
        pos_dic,
        nodelist= p.keys(),
        node_size = node_sizes,
        node_color=list(colors.values()),
        alpha = alphas
    )
    plt.axis("off")
    m.drawcountries(linewidth = 1)
    m.drawstates(linewidth = 0.2)
    m.drawcoastlines(linewidth=1)
    plt.tight_layout()
    
    if save:
        plt.savefig(path, format = "png", dpi = 300)
    plt.tight_layout()
    plt.show()
```
