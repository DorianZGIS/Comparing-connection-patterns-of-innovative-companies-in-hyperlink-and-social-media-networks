```python
import pandas as pd
import networkx as nx
import pickle5 as pickle
import numpy as np
```

```python
haversine_matrix = pd.read_csv('/Users/dorianbenutzer/Desktop/Digital_Twitter_Layer_local/data/graphs/haversine_matrix_with_names.csv')
# The distance is in km
```

```python
with open('/Users/dorianbenutzer/Desktop/Digital_Twitter_Layer_local/data/graphs/pos_dic2_hyp.p', 'rb') as fp:
    pos_dic = pickle.load(fp)
weighted_directed_hyplinks_graph = nx.read_gml("/Users/dorianbenutzer/Desktop/Digital_Twitter_Layer_local/data/graphs/weighted_directed_hyplinks_graph")
weighted_directed_followers_graph = nx.read_gml("/Users/dorianbenutzer/Desktop/Digital_Twitter_Layer_local/data/graphs/weighted_directed_followers_graph")
```

# Create edge_dic

```python
edges = list(weighted_directed_hyplinks_graph.edges())
nodes = list(weighted_directed_hyplinks_graph.nodes())
edges_dic_h = {}

for edge in edges:
    key = edge[0]
    val = edge[1]
    if key not in edges_dic_h.keys():
        edges_dic_h[key] = val
    else:
        val_old = edges_dic_h[key]
        val_new = val_old + ',' + val
        edges_dic_h[key] = val_new
```

```python
edges = list(weighted_directed_followers_graph.edges())
nodes = list(weighted_directed_followers_graph.nodes())
edges_dic_f = {}

for edge in edges:
    key = edge[0]
    val = edge[1]
    if key not in edges_dic_f.keys():
        edges_dic_f[key] = val
    else:
        val_old = edges_dic_f[key]
        val_new = val_old + ',' + val
        edges_dic_f[key] = val_new
```

```python
print(len(edges_dic_f.keys()))
print(len(edges_dic_h.keys()))
```

# Create mean_geographical_proximity_dic

```python
nodes = pos_dic.keys()
haversine_matrix.loc['@SwiftSystemsInc', '@netedfound']
```

```python
haversine_matrix.head()
```

```python
haversine_matrix = haversine_matrix.drop('Unnamed: 0', axis=1)
```

```python
haversine_matrix.index = haversine_matrix.columns
```

```python
mean_geographical_proximity_dic_hyp = {}
nodes = pos_dic.keys()

for comp in edges_dic_h.keys():
    comp_edge_list = edges_dic_h[comp].split(',')
    number = 0
    sum_distance = 0
    for edge in comp_edge_list:
        sum_distance += haversine_matrix.loc[comp,edge]
        number += 1
    mean_distance = sum_distance/number
    mean_geographical_proximity_dic_hyp[comp] = mean_distance
```

```python
mean_geographical_proximity_dic_foll = {}
nodes = pos_dic.keys()

for comp in edges_dic_f.keys():
    comp_edge_list = edges_dic_f[comp].split(',')
    number = 0
    sum_distance = 0
    for edge in comp_edge_list:
        sum_distance += haversine_matrix.loc[comp,edge]
        number += 1
    mean_distance = sum_distance/number
    mean_geographical_proximity_dic_foll[comp] = mean_distance
```

```python
mean_geographical_proximity_dic_foll
```

# Create a dataframe

```python
mgph = []
for node in nodes:
    if node in mean_geographical_proximity_dic_hyp.keys():
        val = mean_geographical_proximity_dic_hyp[node]
    else:
        val = 0 
    mgph.append(val)
mgpf = []
for node in nodes:
    if node in mean_geographical_proximity_dic_foll.keys():
        val = mean_geographical_proximity_dic_foll[node]
    else:
        val = 0
    mgpf.append(val)
```

```python
data = pd.DataFrame(zip(nodes,mgph,mgpf),columns = ['nodes','mean_geographical_proximity_hyp','mean_geographical_proximity_foll'])
```
